```python
import findspark
findspark.init()
findspark.find()
import pyspark
import pyspark
import matplotlib.pyplot as plt
import pandas as pd
from pyspark.sql import SparkSession
```

```python
conf = pyspark.SparkConf() 
sc = pyspark.SparkContext('local','visualization notebook')
spark = SparkSession.builder.getOrCreate()
```

```python
def parseLine(line):
    fields = line.split(',')
    key = str(fields[0][1:])
    count = int(fields[1][:-1])
    return (key,count)
```

```python
textRDD = sc.textFile("/kmer_counting/output_java_spark_3mer/part-00000")
ordersTuple=textRDD.map(parseLine)
print(ordersTuple.count())
```

```python
columns = ['key','count']
ax = ordersTuple.toDF(columns).toPandas().plot.line(x='key', y='count', rot=0, figsize=(9,6))
ax.set_title("3mer")
ax.get_xaxis().set_visible(False)
```

```python
textRDD = sc.textFile("/kmer_counting/output_java_spark_3mer/part-00000")
ordersTuple=textRDD.map(parseLine).take(10000)
print(len(ordersTuple))
```

```python
columns = ['key','count']
ax = pd.DataFrame(ordersTuple, columns=columns).plot.area(y='count', rot=0, figsize=(9,6))
ax.set_title("3mer")
ax.set_xlim([-50,10000])
```
